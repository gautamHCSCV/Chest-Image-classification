```python
import seaborn as sns
import os
import pandas as pd
import numpy as np
import shutil
import warnings
warnings.filterwarnings("ignore")
```

```python
path0 = '../../dataset/Stanford_chestxray_dataset/'
path = '../../dataset/Stanford_chestxray_dataset/CheXpert-v1.0-small/'
os.listdir(path)
```

```python
path1 = path + 'train.csv'
df = pd.read_csv(path1)
df.head(10)
```

```python
len(os.listdir('../../dataset/STANFORD_DATA/stanford_chest_train/normal'))
```

```python
dst_dir = "../../dataset/STANFORD_DATA/stanford_balanced/"
abnorm,norm = 0,0
for i in range(len(df)):
    if df['No Finding'][i]==1.0:
        if norm<22380:
            shutil.copy(path0+df['Path'][i], dst_dir+'normal/{}.jpg'.format(i))
            norm+=1
        continue
    elif abnorm<22380:
        shutil.copy(path0+df['Path'][i], dst_dir+'abnormal/{}.jpg'.format(i))
        abnorm+=1
```

```python
%pylab inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
img = mpimg.imread(path0+df['Path'][0])
imgplot = plt.imshow(img)
plt.show()
```
